```python
# Program to find substring
s1 = input("Enter a string: ")
s2 = input("Enter a substring to find: ")

i = s1.find(s2)

if i == -1:
    print(f"{s2} not found in {s1}")
else:
    print(f"{s2} found at {i}th index of {s1}")
```

```python
# Write a program to make 1st letter of a sentance capital in a below string
s6 = "hello everyone. how are you?"
out_list=[]
splitted_list = s6.split('. ')
for sentance in splitted_list:
    #print(sentance)
    out_list.append(sentance.capitalize())
    #print(out_list)
    
sentence_capa = '. '.join(out_list)
print(sentence_capa)
```
