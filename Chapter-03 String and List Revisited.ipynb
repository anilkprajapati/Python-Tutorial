# Ascii value


```python
# There is a differnce of 32 between small and capital letters
```

```python
s1='a'
ascii1=ord(s1)
print(ascii1)
```

```python
# converting small letter into capital using ascii
ascii2=ascii1-32
print(ascii2)
print(chr(ascii2))
```

```python
# Program to find substring
s1 = input("Enter a string: ")
s2 = input("Enter a substring to find: ")

i = s1.find(s2)

if i == -1:
    print(f"{s2} not found in {s1}")
else:
    print(f"{s2} found at {i}th index of {s1}")
```
